```python
import pymorphy2
from fuzzywuzzy import fuzz
```

```python
df = pd.read_csv('медицинские_протоколы.csv', sep=';')
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
df['Пол'] = df['Пол'].fillna(df['Пол'].value_counts().keys()[0])
```

```python
df
```

```python
complaints = pd.DataFrame()
```

```python
all_complaints = list()
```

```python
ma = -1
for i in df['Жалобы']:
  ma = max(ma, len(i.split(',')))
  ans = []
  for j in i.split(','):
    ans.append(j.rstrip().lstrip().lower())
  all_complaints.extend(ans)
```

```python
ma
```

```python
complaints['comp'] = all_complaints
```

```python
complaints['comp'].value_counts().head(11)
```

```python
complaints
```

```python
dict_complaints = {}
```

```python
for i in complaints['comp']:
  flag = True
  for q in dict_complaints.keys():
    if fuzz.ratio(q, i) >= 60:
      flag = False
      dict_complaints[q][1].add(i)
      dict_complaints[q][0] += 1
      break
  else:
    dict_complaints[i] = [1, set([i])]
```

```python
dict_complaints
```

```python
jalobs = list(dict_complaints.values())
```

```python
jalobs.sort(key=lambda x: x[0])
```

```python
import pickle
```

```python
with open('jalobs.pkl', 'wb') as f:
    pickle.dump(jalobs, f)
```
